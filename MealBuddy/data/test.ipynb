```python
import pandas as pd
import numpy as np
import re
```

```python
# Load the CSV file
df = pd.read_csv("nutrition.csv") 
```

```python
display(df)
```

```python
df.dtypes
```

```python
# Replace missing values with 0
df.fillna(0, inplace=True)
```

```python
# Remove the column if it exists
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
display(df)
```

```python
# Remove letters from numeric columns and convert them to float
numeric_columns = [
    "serving_size", "calories", "total_fat", "saturated_fat", "cholesterol", "sodium",
    "choline", "folate", "folic_acid", "niacin", "pantothenic_acid", "riboflavin", "thiamin",
    "vitamin_a", "vitamin_a_rae", "carotene_alpha", "carotene_beta", "cryptoxanthin_beta",
    "lutein_zeaxanthin", "lucopene", "vitamin_b12", "vitamin_b6", "vitamin_c", "vitamin_d",
    "vitamin_e", "tocopherol_alpha", "vitamin_k", "calcium", "copper", "irom", "magnesium",
    "manganese", "phosphorous", "potassium", "selenium", "zink", "protein", "alanine",
    "arginine", "aspartic_acid", "cystine", "glutamic_acid", "glycine", "histidine",
    "hydroxyproline", "isoleucine", "leucine", "lysine", "methionine", "phenylalanine",
    "proline", "serine", "threonine", "tryptophan", "tyrosine", "valine", "carbohydrate",
    "fiber", "sugars", "fructose", "galactose", "glucose", "lactose", "maltose", "sucrose",
    "fat", "saturated_fatty_acids", "monounsaturated_fatty_acids", "polyunsaturated_fatty_acids",
    "fatty_acids_total_trans", "alcohol", "ash", "caffeine", "theobromine", "water"
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).apply(lambda x: re.sub(r"[^\d.]", "", x))  # Remove non-numeric characters
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to float
```

```python
display(df)
```

```python
df.dtypes
```

```python
# Save the cleaned data for importing into the app
df.to_csv("cleaned_nutrition.csv", index=False)
```

```python
# Export to JSON
df.to_json("cleaned_nutrition.json", orient="records", indent=4)
```
