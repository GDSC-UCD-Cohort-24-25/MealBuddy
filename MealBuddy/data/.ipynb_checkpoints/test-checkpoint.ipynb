```python

import pandas as pd
import requests
import time
import os

API_KEY = "yO7AbLAPffMuhiGQp4QWgn3uyvycpa5lMgj637oy"  

classified_df = pd.read_csv("classified_nutrition.csv")

def get_food_group(food_name, api_key):
    """Query the USDA FoodData Central API to get food group information."""
    url = "https://api.nal.usda.gov/fdc/v1/foods/search"
    params = {
        "api_key": api_key,
        "query": food_name,
        "dataType": ["Foundation", "SR Legacy"],
        "pageSize": 1
    }
    try:
        response = requests.get(url, params=params)
        if response.status_code == 200:
            data = response.json()
            if data.get("foods"):
                return data["foods"][0].get("foodCategory", "Unknown")
        return "Unknown"
    except Exception as e:
        print(f"Error for '{food_name}': {e}")
        return "Unknown"

# Only classify rows that are missing or have "Unknown" as food group
mask = classified_df["food_group"].isna() | (classified_df["food_group"] == "Unknown")
unclassified = classified_df[mask].copy()

for idx, row in unclassified.iterrows():
    food_name = row["name"]
    group = get_food_group(food_name, API_KEY)
    classified_df.at[idx, "food_group"] = group
    print(f"Classified '{food_name}' as '{group}'")
    time.sleep(1)  # Respect API rate limit
```

```python
import re

# Function to clean food names
def clean_name(name):
    name = name.lower()
    name = re.sub(r'[\\"\'\[\]\(\)/,.:;!?{}@#$%^&*<>+=~`|]', '', name)
    name = re.sub(r'\s+', ' ', name)
    return name.strip()

# Filter only the unknowns
retry_mask = classified_df["food_group"] == "Unknown"
retry_df = classified_df[retry_mask].copy()

# Retry classification
for idx, row in retry_df.iterrows():
    original_name = row["name"]
    cleaned_name = clean_name(original_name)

    group = get_food_group(cleaned_name, API_KEY)
    classified_df.at[idx, "food_group"] = group
    print(f"Reclassified '{original_name}' → '{group}'")

    time.sleep(0.4)  # respect API rate limit

# Save final result
classified_df.to_csv("classified_nutrition_final.csv", index=False)
print("Retry complete. Saved to 'classified_nutrition_final.csv'")
```

```python
# images

import pandas as pd
import requests
import time
import os

CSV_PATH = "classified_nutrition_final.csv"
OUTPUT_PATH = "image_enriched_nutrition.csv"

# Use the updated file if it exists
df = pd.read_csv(OUTPUT_PATH if os.path.exists(OUTPUT_PATH) else CSV_PATH)

BING_API_KEYS = [
    "YOUR_FIRST_BING_KEY",
    "YOUR_SECOND_BING_KEY",
    # Add more keys here
]
api_index = 0

HEADERS = lambda key: {
    "Ocp-Apim-Subscription-Key": key
}

def fetch_bing_image(query, api_key):
    search_url = "https://api.bing.microsoft.com/v7.0/images/search"
    params = {
        "q": f"{query} food",
        "license": "public",
        "imageType": "photo",
        "count": 1
    }
    try:
        response = requests.get(search_url, headers=HEADERS(api_key), params=params)
        if response.status_code == 429:
            return "RATE_LIMITED"
        response.raise_for_status()
        data = response.json()
        if "value" in data and data["value"]:
            return data["value"][0]["contentUrl"]
        else:
            return None
    except Exception as e:
        print(f"Error for '{query}': {e}")
        return None

updated = 0

for i, row in df.iterrows():
    # Skip rows that already have valid images
    current_url = str(row.get("image_url", ""))
    if current_url and "dummyimage" not in current_url and not pd.isna(current_url):
        continue

    food_name = row["name"]
    while True:
        current_key = BING_API_KEYS[api_index]
        image_url = fetch_bing_image(food_name, current_key)

        if image_url == "RATE_LIMITED":
            api_index += 1
            if api_index >= len(BING_API_KEYS):
                print("❌ All keys exhausted. Saving progress and exiting.")
                df.to_csv(OUTPUT_PATH, index=False)
                exit()
            print(f"⚠️ Rate limit hit. Switching to key #{api_index + 1}")
            time.sleep(1)
            continue
        break

    if image_url:
        df.at[i, "image_url"] = image_url
        updated += 1
        print(f"[{updated}] ✅ {food_name}: Image added.")
    else:
        print(f"[{i}] ❌ {food_name}: No image found.")

    # Save progress after every image
    df.to_csv(OUTPUT_PATH, index=False)

print(f"✅ Done. {updated} images added.")
```
